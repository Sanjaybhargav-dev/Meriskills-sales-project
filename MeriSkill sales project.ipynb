```python
monthly_sales = df.groupby('Month')['Sales'].sum()
```

```python
top_products = df.groupby('Product')['Quantity Ordered'].sum().sort_values(ascending=False).head(10)
```

```python
top_products
```

```python
city_sales = df.groupby('City')['Sales'].sum().sort_values(ascending=False)
```

```python
city_sales
```

```python
hourly_sales = df.groupby('Hour')['Sales'].sum()
```

```python
hourly_sales
```

```python
#monthly sales plot:

plt.figure(figsize=(10, 6))
plt.bar(monthly_sales.index, monthly_sales.values)
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.title('Monthly Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
#top product Bar chart:

top_products.plot(kind='bar')
plt.xlabel('Product')
plt.ylabel('Total Quantity Sold')
plt.title('Top Selling Products')
plt.show()
```

```python
#City Sales Pie Chart:

city_sales.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.title('City-wise Sales Distribution')
plt.show()
```

```python
#hourly sales line char:

hourly_sales.plot(kind='line', marker='o')
plt.xlabel('Hour')
plt.ylabel('Total Sales ($)')
plt.title('Hourly Sales Trends')
plt.grid(True)
plt.show()
```
